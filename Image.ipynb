```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python

image = cv.imread('1516551.jpg')

# grayscale
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

histogram_image = cv.calcHist([image], [0], None, [256], [0, 256])

histogram_gray_image = cv.calcHist([gray_image], [0], None, [256], [0, 256])
# plot
plt.subplot(2, 2, 1)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.plot(histogram_image, color='red')
plt.title('Histogram - Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.plot(histogram_gray_image, color='black')
plt.title('Histogram - Grayscale Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
